# Feature Engineering
Adding new feature for sqft_lot15.

Exploring sqft_lot15 feature

```python
df.sqft_lot15.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['sqft_lot15'].apply(is_float)].head(10)
```

Above shows that sqft_lot15 can be a range (e.g. 5000 - 10000). For such case we can just
take average of min and max value in the range. There are other cases such as 34.46Sq.
Meter which one can convert to square ft using unit conversion. I am going to just drop
such corner cases to keep things simple

```python
def convert_sqft_to_num(x):
    tokens = x.split('-') # split the range values into two values
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2 # this will return avg
    try:
        return float(x) # this will return the converted avg value into float
    except:
        return None
```

```python
loc_stats = df['zipcode'].value_counts().sort_values(ascending=False)
```

```python
loc_stats
```

```python
loc_less_10 = loc_stats[loc_stats<10]
```

```python
# if the occurence of location is less than 10 make it other
df['zipcode'] = df['zipcode'].apply(lambda x:'other' if x in loc_less_10 else x)
```
